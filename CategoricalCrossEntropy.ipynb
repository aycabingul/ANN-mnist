```
import numpy as np

#hesaplanan degerler, Y
p1 = np.array([1.0, 0, 0])
p2 = np.array([0, 1.0, 0])
p3 = np.array([1.0, 0, 0])

#olmasi istenen degerler, T
t1 = np.array([0, 1.0, 0])
t2 = np.array([0, 0, 1.0])
t3 = np.array([1.0, 0, 0])

eps = 1e-10 #log(0) undefined oldugu icin

ce = -(np.matmul(t1, np.log(p1+eps)) + np.matmul(t2, np.log(p2+eps)) + np.matmul(t3, np.log(p3+eps)))

print("categorical cross entropy =", ce)
```
